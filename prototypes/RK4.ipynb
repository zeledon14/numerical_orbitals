```python
def y(x):
    return np.exp(x**2.0)

def yp(x,y):
    return 2.0*x*y
```

```python
#rk4 dev 
h=0.001
grid= np.arange(0.0, 2.0, h)
```

```python
y_func= y(grid)
```

```python
w=[y_func[0]]
```

```python
#RK4
for i, xi in enumerate(grid):
    k=4*[0.0]
    for j in range(4):
        if j == 0:
            k[j]= h*yp(xi,w[i])
        elif j == 1 or j == 2:
            k[j]= h*yp(xi+ 0.5*h,w[i] + 0.5*k[j-1])
        else:
            k[j]= h*yp(xi+h, w[i]+k[j-1])
    w.append(w[i] + (1.0/6.0)*(k[0] + 2.0*k[1] + 2.0*k[2] + k[3]))
```

```python
plt.plot(grid,np.array(w[:-1]),'r+', grid,y_func,'b+')
plt.show()
```
