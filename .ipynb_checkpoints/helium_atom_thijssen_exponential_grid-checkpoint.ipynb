```python
def get_ue_hydrogen(grid):
    r= np.array(grid)
    return -1.0*(np.multiply(np.add(r,1.0),np.exp(-2.0*r))) +1
ue_hydrogen=get_ue_hydrogen(exp_grid)
v_hart_hydro= np.divide(ue_hydrogen,exp_grid)
```
