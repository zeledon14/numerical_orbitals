```python
import numpy as np
import matplotlib.pyplot as plt
import json
import os
plt.rcParams["figure.figsize"] = (25,10)

plt.rcParams.update({'font.size': 22})

colors='bgrykcm'
symbol='*-+^'
```

```python
import Numerov
import utils
import Hartree_XC
import numpy as np
import Atomic_Like_Potential as ALP
from Radial_Eigenfunction import Radial_Eigenfunction
```

```python
#defining angular momentum and energy grids to look for eigenfunctions
angu_mome_grid=[0]
r_N_grid=[10.0]
#initial energy guess
ener_grid= -1.0*np.array(utils.get_exponential_grid(delta=0.001, max_valu=2.5, grid_points=10))
#ener_grid= -1.0*np.flip(np.arange(0.01, 2.2, 0.13, dtype=float))
#here we are only interested in the base state
#ener_grid= ener_grid[-3:]#[-0.4,-0.6]
ener_grid
```

```python
#start kwargs with default values
kwargs={'delta_x': 0.01, 'r_N':10.0, 'l':0, 'Z':2.0, 'E':-0.5}
#inittial values of u function for integration routine
y_func=[2.5e-6, 5.0e-6]
y_func_p=[r_N_grid[0]*np.exp(-1.0*r_N_grid[0]),
        (r_N_grid[0]-kwargs['delta_x'])*np.exp(-1.0*(r_N_grid[0]-kwargs['delta_x']))]
y_func= y_func_p
```

```python
y_func_p
```

```python
#for the r_N and angu_mome grids calcuate u(0,E)
#find the limits of the nodes for u(0,E) as function of E 
#the nodes limits are Ea and Eb. The eigenvalue E follows Ea < E < Eb
pote_eige_func=[]
for r_N in r_N_grid:
    kwargs['r_N']= r_N
    r_grid_reversed= utils.get_uniform_r_grid(**kwargs)
    S_atomic_radial= len(r_grid_reversed)*[0.0e-8]
    VH_inde_elec=len(r_grid_reversed)*[0.0e-8]
    for l in angu_mome_grid:
        kwargs['l']= l
        
        u0_e=[]
        for E in ener_grid:
            kwargs['E']=E
            #print(kwargs)
            u_func, r_grid= Numerov.Numerov_backwards(y_func, r_grid_reversed,VH_inde_elec, 
                                                      ALP.K2_atomic_radial,
                                                      S_atomic_radial,**kwargs)
            #print(u_func[0])
            u0_e.append(u_func[0])
        number_of_nodes, nodes_positions= utils.get_nodes_information(u0_e, ener_grid)
        #print(number_of_nodes)
        if number_of_nodes > 0:
            for elem in nodes_positions:
                pote_eige_func.append(Radial_Eigenfunction(Ea=elem[0],Eb=elem[1],**kwargs))
```

```python
#For every potential eigenfunction shrink the Ea, Eb segment
if len(pote_eige_func) > 0:
    for radi_eige_func in pote_eige_func:
        kwargs={'delta_x':radi_eige_func.delta_x, 'r_N':radi_eige_func.r_N, 
                'l':radi_eige_func.l, 'Z':radi_eige_func.Z, 'E':radi_eige_func.Ea}
        r_grid_reversed= utils.get_uniform_r_grid(**kwargs)
        while not radi_eige_func.check_if_eigenvalue():
            
            
            kwargs={'delta_x':radi_eige_func.delta_x, 'r_N':radi_eige_func.r_N, 
                    'l':radi_eige_func.l, 'Z':radi_eige_func.Z, 'E':radi_eige_func.Ea}
            
            u_func, r_grid= Numerov.Numerov_backwards(y_func, r_grid_reversed,VH_inde_elec, 
                                                      ALP.K2_atomic_radial,
                                                      S_atomic_radial,**kwargs)
            u0a= u_func[0]
            #print('u0a ', u0a, ' Ea ', radi_eige_func.Ea)
            kwargs={'delta_x':radi_eige_func.delta_x, 'r_N':radi_eige_func.r_N, 
                    'l':radi_eige_func.l, 'Z':radi_eige_func.Z, 'E':radi_eige_func.Eb}
            
            u_func, r_grid= Numerov.Numerov_backwards(y_func, r_grid_reversed,VH_inde_elec, 
                                                      ALP.K2_atomic_radial,
                                                      S_atomic_radial,**kwargs)
            u0b= u_func[0]
            #print('u0b ', u0b, ' Eb ', radi_eige_func.Eb)
            #E' is the candidate energy to shrink the energy segment
            Ep= utils.false_position(radi_eige_func.Ea,
                                     radi_eige_func.Eb,
                                     u0a,u0b)
            kwargs={'delta_x':radi_eige_func.delta_x, 'r_N':radi_eige_func.r_N, 
                    'l':radi_eige_func.l, 'Z':radi_eige_func.Z, 'E':Ep}
            u_func, r_grid= Numerov.Numerov_backwards(y_func, r_grid_reversed,VH_inde_elec, 
                                                      ALP.K2_atomic_radial,
                                                      S_atomic_radial,**kwargs)
            u0p= u_func[0]
            #print('u0p ', u0p, ' Ep ', Ep)
            if np.sign(u0p) == np.sign(u0a):
                radi_eige_func.Ea=Ep
            else:# np.sign(U0p) == np.sign(U0b):
                radi_eige_func.Eb=Ep
            if abs(u0p) < 1e-18:
                radi_eige_func.Ea=Ep
                radi_eige_func.Eb=Ep
        print(radi_eige_func.E)
        print('********************************')
    for radi_eige_func in pote_eige_func:
        if radi_eige_func.E != None:
            kwargs={'delta_x':radi_eige_func.delta_x, 'r_N':radi_eige_func.r_N, 
                    'l':radi_eige_func.l, 'Z':radi_eige_func.Z, 'E':radi_eige_func.E}
            r_grid_reversed= utils.get_uniform_r_grid(**kwargs)
            u_func, r_grid= Numerov.Numerov_backwards(y_func, r_grid_reversed,VH_inde_elec, 
                                                      ALP.K2_atomic_radial,
                                                      S_atomic_radial,**kwargs)
            radi_eige_func.set_u_function(u_func)
            radi_eige_func.set_r_grid(r_grid)
            radi_eige_func.bild_normalized_radi_eige_func()
```

```python
for i, radi_eige_func in enumerate(pote_eige_func):
    colo_symb=colors[i%len(colors)]+symbol[i%len(symbol)]
    label=f'index {i} E_{str(radi_eige_func.E)[:7]} Z_{radi_eige_func.Z}_l_{radi_eige_func.l}'
    plt.plot(radi_eige_func.r_grid,radi_eige_func.u_func**2,colo_symb,  label=label)
plt.legend()
plt.show()
```

```python
r_grid_reversed= utils.get_uniform_r_grid(**kwargs)
u_func_init=y_func_p#[1.8e-8, 2.0e-8]
S_atomic_radial= len(r_grid_reversed)*[0.0e-8]
#uH_func_init=[kwargs['Z'],kwargs['Z']]
uH_func_init_forw= [0.0, kwargs['delta_x']]

#radi_eige_func.u_func= (2.0**0.5)*radi_eige_func.u_func#2 electrons in orbital


syst_E=2.0 #start system Energy value to an unrealistic value
syst_Ep=-1.0
while abs(syst_E- syst_Ep) > 10e-7:
    #plt.plot(radi_eige_func.r_grid,radi_eige_func.u_func**2.0)
    #plt.show()
    #print('u_func(0) -> ',radi_eige_func.u_func[0])
    syst_E= syst_Ep #setting system energy for next try
    S_uH= Hartree_XC.S_Hartree_for_poisson_equation(radi_eige_func.r_grid, radi_eige_func.u_func)
    S_uH_frwd= Hartree_XC.S_Hartree_for_poisson_equation(radi_eige_func.r_grid, radi_eige_func.u_func, Numerov_backwards=False)
    #numerov solving poisson equation for electrostatic potential
#    uH_func, r_grid= Numerov.Numerov_backwards(uH_func_init, r_grid_reversed, VH_inde_elec,
#                                              Hartree_XC.K2_Hartree_XC_for_poisson_equation,
#                                              S_uH,**kwargs)
#    uH_func= Hartree_XC.uH_func_boundery_condition(uH_func, r_grid, kwargs['Z'])
    uH_func_fwrd, r_grid= Numerov.Numerov_forward(uH_func_init_forw, r_grid, VH_inde_elec,
                                              Hartree_XC.K2_Hartree_for_poisson_equation,
                                              S_uH_frwd,**kwargs)
    #print('uH_func_fwrd[0] before baundery condition', uH_func_fwrd[0])
    uH_func= Hartree_XC.uH_func_boundery_condition_forward(uH_func_fwrd, r_grid, 1.00)#kwargs['Z'])
    #print('uH_func[0] after baundery condition', uH_func[0])
    #The electrostratic potential is fliped to be input into the Numerov.Numerov_backwards
    #for solving schrodinger equation with it
    VH_pote= np.flip(np.divide(uH_func,r_grid))
    Vxc_pote= np.flip(Hartree_XC.get_Vxc(r_grid, radi_eige_func.u_func))
    V_inte_pote= list(VH_pote - Vxc_pote)
# find where the new energy nodes bouderies are
    u0_e=[]
    for E in ener_grid:
        kwargs['E']=E
        #print(kwargs)
        u_func, r_grid= Numerov.Numerov_backwards(u_func_init, r_grid_reversed, V_inte_pote,
                                              Hartree_XC.K2_SE_VH,
                                              S_atomic_radial,**kwargs)
        #print(u_func[0])
        u0_e.append(u_func[0])
    number_of_nodes, nodes_positions= utils.get_nodes_information(u0_e, ener_grid)
    #print('nodes_positions ', nodes_positions)
    min_indx= np.argsort([elem[0] for elem in nodes_positions])[0] #index of minimal energy window
    radi_eige_func.Ea=nodes_positions[min_indx][0]
    radi_eige_func.Eb=nodes_positions[min_indx][1]
#use the current energy window to find the new eigenvalue
    while not radi_eige_func.check_if_eigenvalue():


        kwargs={'delta_x':radi_eige_func.delta_x, 'r_N':radi_eige_func.r_N, 
                'l':radi_eige_func.l, 'Z':radi_eige_func.Z, 'E':radi_eige_func.Ea}

        u_func, r_grid= Numerov.Numerov_backwards(u_func_init, r_grid_reversed, V_inte_pote,
                                              Hartree_XC.K2_SE_VH,
                                              S_atomic_radial,**kwargs)
        u0a= u_func[0]
        #print('u0a ', u0a)
        kwargs={'delta_x':radi_eige_func.delta_x, 'r_N':radi_eige_func.r_N, 
                'l':radi_eige_func.l, 'Z':radi_eige_func.Z, 'E':radi_eige_func.Eb}

        u_func, r_grid= Numerov.Numerov_backwards(u_func_init, r_grid_reversed, V_inte_pote,
                                              Hartree_XC.K2_SE_VH,
                                              S_atomic_radial,**kwargs)
        u0b= u_func[0]
        #print('u0b ', u0b)
        #E' is the candidate energy to shrink the energy segment
        Ep= utils.false_position(radi_eige_func.Ea,
                                 radi_eige_func.Eb,
                                 u0a,u0b)
        kwargs={'delta_x':radi_eige_func.delta_x, 'r_N':radi_eige_func.r_N, 
                'l':radi_eige_func.l, 'Z':radi_eige_func.Z, 'E':Ep}
        u_func, r_grid= Numerov.Numerov_backwards(u_func_init, r_grid_reversed, V_inte_pote,
                                              Hartree_XC.K2_SE_VH,
                                              S_atomic_radial,**kwargs)
        u0p= u_func[0]
        #print('u0p ', u0p)
        #print('Ep ', Ep)
        #substitute Ep for either Ea or Eb
        if np.sign(u0p) == np.sign(u0a):

            radi_eige_func.Ea=Ep
        else:# np.sign(U0p) == np.sign(U0b):

            radi_eige_func.Eb=Ep
        if abs(u0p) < 1e-20:
            radi_eige_func.Ea=Ep
            radi_eige_func.Eb=Ep
#once the new eigenvalue is found calculate new 
    kwargs={'delta_x':radi_eige_func.delta_x, 'r_N':radi_eige_func.r_N, 
            'l':radi_eige_func.l, 'Z':radi_eige_func.Z, 'E':radi_eige_func.E}
    u_func, r_grid= Numerov.Numerov_backwards(u_func_init, r_grid_reversed, V_inte_pote,
                                          Hartree_XC.K2_SE_VH,
                                          S_atomic_radial,**kwargs)
    radi_eige_func.set_u_function(u_func)
    radi_eige_func.set_r_grid(r_grid)
    radi_eige_func.bild_normalized_radi_eige_func()
    
        
    #calculate new syst_Ep

    #electrostatic energy
    VH_pote_for_E= np.flip(VH_pote) #flip to be in same order as u_func
    Vxc_pote_for_E= np.flip(Vxc_pote) #flip to be in same order as u_func
    VH_E= utils.integrate_f(VH_pote_for_E*(radi_eige_func.u_func)**2.0,radi_eige_func.delta_x)
    Vxc_E= utils.integrate_f(Vxc_pote_for_E*(radi_eige_func.u_func)**2.0,radi_eige_func.delta_x)
    syst_Ep= 2.0*radi_eige_func.E - 2.0*VH_E + 0.5*Vxc_E
    print('syst_Ep ', syst_Ep)
    print('radi_eige_func.E ', radi_eige_func.E)
    print('VH_E ', VH_E)
    print('Vxc_E ', Vxc_E)
    print('**************')
```
