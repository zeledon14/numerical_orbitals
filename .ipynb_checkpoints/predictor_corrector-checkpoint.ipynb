```python
# RK4 implementation

E= kwargs['E']
h= kwargs['delta_x']
Z= kwargs['Z']
l= kwargs['l']
w10= u_hydr[0]
w20= (u_hydr[1] - u_hydr[0])/h

w1=[w10]
w2=[w20]

for i,ri in enumerate(r_grid_reversed):
    k1=4*[0.0]
    k2=4*[0.0]
    v0= ALP.V_coulomb(ri, Z) + ALP.V_angular(ri, l)
    v0p= ALP.V_coulomb(ri + 0.5*h, Z) + ALP.V_angular(ri + 0.5*h, l)
    for j in range(4):
        #k1[j]=h*(w2[i] + 0.5*k2[j-1])
        if j == 0:
            k1[j]=h*(w2[i])
            k2[j]= h*2.0*(v0 - E)*w1[i]
        elif j == 1 or j == 2:
            k1[j]=h*(w2[i] + 0.5*k2[j-1])
            k2[j]= h*2.0*(v0p - E)*(w1[i] + 0.5*k1[j-1])
        elif j == 3:
            k1[j]=h*(w2[i] + k2[j-1])
            k2[j]= h*2.0*(v0p- E)*(w1[i] + k1[j-1])
    w1.append( w1[i] + (1.0/6.0)*(k1[0] + 2.0*k1[1] + 2.0*k1[2] + k1[3]))
    w2.append( w2[i] + (1.0/6.0)*(k2[0] + 2.0*k2[1] + 2.0*k2[2] + k2[3]))
```

```python
plt.plot(r_grid_reversed,np.array(w1[:-1])**2,'r+', r_grid_reversed,u_hydr**2,'b+')
plt.show()
```
